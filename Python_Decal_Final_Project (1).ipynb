```python
import pandas as pd
import matplotlib.pyplot as plt

#PLANETS WITH ESI>0.5 GRAPH 
df = pd.read_csv(r"C:\Users\cadyw\Downloads\ExoplanetDatafr - Sheet1.csv")
x = df[df['ESI'] >= 0.5]['ESI'].tolist()
y = df[df['ESI'] >= 0.5]['Distance'].tolist()

fig=plt.figure()
ax=fig.add_axes([0, 0, 1, 1])

ax.set_xlabel('Exoplanet Ideal ESI', fontsize = 14)
ax.set_ylabel('Distance From Exoplanet Sun', fontsize = 14)
ax.set_title('Ideal Planets')
plt.ylim(0, 3000)
plt.plot(x, y, 'ok')

#ALL PLANETS GRAPH
df = pd.read_csv(r"C:\Users\cadyw\Downloads\ExoplanetDatafr - Sheet1.csv")
x = df['ESI'].tolist()
y = df['Distance'].tolist()

fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])

ax.set_xlabel('All Exoplanet ESI', fontsize = 14)
ax.set_ylabel('All Distance From Exoplanet Sun', fontsize = 14)
ax.set_title('All Exoplanets in Database')
plt.ylim(0, 3000)

plt.plot(x, y, 'ok')
```
