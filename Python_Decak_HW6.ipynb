```python
import numpy as np
import matplotlib.pyplot as plt

def sinusoid(x, A, w):
    return A * np.sin(w * x)

x = np.linspace(0, 2 * np.pi, 1000)
A_values = np.array([1, 2, 3, 4, 5])
w_values = np.array([1, 2, 3, 4, 5]) 
plt.figure(figsize=(10, 6)) 

for A in A_values:
    for w in w_values:
        y = sinusoid(x, A, w)  # Calculate y data for the current A and w
        plt.plot(x, y, label=f'A={A}, w={w}')

plt.legend()
plt.xlabel('x')
plt.ylabel('y = A * sin(w * x)')
plt.title('Sinusoidal Waves for Different A and w Values')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

data1 = np.random.uniform(0, 100, 40)
data2 = np.random.uniform(0, 100, 40) 


plt.figure(figsize=(10, 6))

plt.plot(data1, color='orange', linewidth=10, label='Data 1 (Orange)')

plt.plot(data2, color='red', linestyle='--', label='Data 2 (Red Dashed)')

plt.xlabel('Index')
plt.ylabel('Random Value')
plt.title('Plot of Random Data')

plt.legend()

plt.show()
```
